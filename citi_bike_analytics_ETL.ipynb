# Citi Bike Data 2018-2019

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
path_data_01 = "Resources/JC-201901-citibike-tripdata.csv"
path_data_02 = "Resources/JC-201902-citibike-tripdata.csv"
path_data_03 = "Resources/JC-201903-citibike-tripdata.csv"
path_data_04 = "Resources/JC-201904-citibike-tripdata.csv"
path_data_05 = "Resources/JC-201905-citibike-tripdata.csv"
path_data_06 = "Resources/JC-201906-citibike-tripdata.csv"
path_data_07 = "Resources/JC-201907-citibike-tripdata.csv"
path_data_08 = "Resources/JC-201908-citibike-tripdata.csv"
path_data_09 = "Resources/JC-201909-citibike-tripdata.csv"
path_data_10 = "Resources/JC-201910-citibike-tripdata.csv"
path_data_11 = "Resources/JC-201911-citibike-tripdata.csv"
path_data_12 = "Resources/JC-201912-citibike-tripdata.csv"
```

```python
# Reading csv ->df
data_01_df=pd.read_csv(path_data_01)
data_02_df=pd.read_csv(path_data_02)
data_03_df=pd.read_csv(path_data_03)
data_04_df=pd.read_csv(path_data_04)
data_05_df=pd.read_csv(path_data_05)
data_06_df=pd.read_csv(path_data_06)
data_07_df=pd.read_csv(path_data_07)
data_08_df=pd.read_csv(path_data_08)
data_09_df=pd.read_csv(path_data_09)
data_10_df=pd.read_csv(path_data_10)
data_11_df=pd.read_csv(path_data_11)
data_12_df=pd.read_csv(path_data_12)
```

```python
#Combine the dataframes
frames=[data_01_df,data_02_df,data_03_df,data_04_df,data_05_df,data_06_df,
       data_07_df,data_08_df,data_09_df,data_10_df,data_11_df,data_12_df]

data_2019_df=pd.concat(frames)
```

```python
#Display the combined df
data_2019_df.head()
```

```python
#Count of the combined df
data_2019_df.count()
```

```python
#Check the data types
data_2019_df.dtypes
```

```python
#Convert the datatypes for starttime and stoptime fileds
data_2019_df["starttime"] = pd.to_datetime(data_2019_df["starttime"])
data_2019_df["stoptime"] = pd.to_datetime(data_2019_df["stoptime"])
data_2019_df.dtypes
```

```python
data_2019_df.head()
```

```python
#Rename columns for readabilty
data_2019_df.columns=["trip_duration","start_time","stop_time",
                      "start_station_id","start_station_name",
                      "start_station_latitude","start_station_longitude",
                      "end_station_id","end_station_name",
                      "end_station_latitude","end_station_longitude",
                     "bike_id","usertype","birth_year","gender"]
data_2019_df.dtypes
```

```python
# In the gender column replace 0-"Unknown", 1-"Male",2-"Female"
data_2019_df=data_2019_df.replace(0, "Unknown")
```

```python
data_2019_df=data_2019_df.replace(1, "Male")
```

```python
data_2019_df=data_2019_df.replace(2, "Female")
```

```python
#Export the file and save as .csv
data_2019_df.to_csv("Resources/2019.csv",index=False,header=True,encoding='utf-8-sig')
data_2019_df.to_csv("2019.csv",index=False,header=True,encoding='utf-8-sig')
```
